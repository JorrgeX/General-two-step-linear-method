```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compute_RAS(alpha0, alpha1, beta_1, beta0, beta1, theta_values):
    rho = lambda z: z**2 + alpha0*z + alpha1
    sigma = lambda z: beta_1*z**2 + beta0*z + beta1

    RAS_values = []

    for theta in theta_values:
        z = np.exp(1j * theta)
        omega = rho(z) / sigma(z)
        RAS_values.append(omega)

    return RAS_values
```

```python
def plot_RAS(alpha_range, figure_number):
    plt.figure(figure_number)
    legend_labels = []

    theta_values = np.linspace(0, 2 * np.pi, 1000)

    for alpha0 in alpha_range:
        alpha1 = -(1 + alpha0)
        beta_1 = 1/3 - alpha0/12
        beta0 = 4/3 + 2*alpha0/3
        beta1 = 1/3 + 5*alpha0/12
        RAS_values = compute_RAS(alpha0, alpha1, beta_1, beta0, beta1, theta_values)
        plt.plot(np.real(RAS_values), np.imag(RAS_values))
        legend_labels.append(f'Î± = {alpha0:.1f}')

    plt.title(f'RAS Boundaries')
    plt.legend(legend_labels)
    plt.grid()
    plt.show()
```

```python
# Plot for Î± = -1.8, -1.7, ..., -1.1
alpha_range_1 = np.arange(-1.8, -1.09, 0.1)
plot_RAS(alpha_range_1, figure_number=1)

# Plot for Î± = -1.0, -0.9, ..., -0.1
alpha_range_2 = np.arange(-1.0, 0, 0.1)
plot_RAS(alpha_range_2, figure_number=1)
```

```python
def check_Root_Condition_1(alpha_range):
    result = {}
    
    for alpha0 in alpha_range:
        alpha1 = -(1 + alpha0)
        beta_1 = 1/3 - alpha0/12
        beta0 = 4/3 + 2*alpha0/3
        beta1 = 1/3 + 5*alpha0/12
        z = np.exp(-0.2 + 0j)
        omega = (z**2 + alpha0 * z + alpha1) / (beta_1 * z**2 + beta0 * z + beta1)
        result[f'Î± = {alpha0:.1f}'] = "Inside" if np.abs(omega) < 1 else "Outside"

    return result
```

```python
result_1 = check_Root_Condition_1(alpha_range_1)

print("Root Condition Check for Î± = -1.8, -1.7, ..., -1.1:")
for key, value in result_1.items():
    print(f'{key}: RAS is {value}')
```
